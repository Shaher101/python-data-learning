```python
df2 = pd.read_csv("kiwilytics_orders.csv")
df2.info()
df2.describe()
print(df2.shape)
```

```python
df2.head()
```

```python
T_Revenue = df2["quantity"] * df2["unit_price"].sum()
print(T_Revenue)
```

```python
df2['total_price'] = df2["quantity"] * df2["unit_price"]
```

```python
top_product = (df2.groupby('product')["quantity"].sum())
print(top_product)
print(60*"*")

top_customer = (df2.groupby('customer_name')["total_price"].sum().sort_values(ascending=False))
print(top_customer)
```

```python
top_customer.head(5).plot(kind= "bar")
```

```python
A = [10,20,30,40,50]
B1 = [100,120,130,140,150]
B2 = [100,150,160,170,180]


plt.plot(A,B1)
plt.plot(A,B2)
plt.title("Chart 2")
plt.xlabel("Size(m2)")
plt.ylabel("Price($)")
plt.legend(['Data 1','Data 2'])
plt.show()
```

```python
plt.plot(A,B1,color = 'blue', marker = "*", linestyle = '--') 
plt.plot(A,B2,color = 'green', marker = "o", linestyle = '-.')
plt.title("Chart 3")
plt.xlabel("Size(m2)")
plt.ylabel("Price($)")
plt.legend(['Data 1','Data 2'])
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(A,B1,color = 'blue', marker = "*", linestyle = '--') 
plt.plot(A,B2,color = 'green', marker = "o", linestyle = '-.')

plt.title("Chart 3")
plt.xlabel("Size(m2)")
plt.ylabel("Price($)")
plt.legend(['Data 1','Data 2'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
